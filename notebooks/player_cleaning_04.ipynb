```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_columns', None)
```

```python
folder_path = '/Users/samielzengotita/Documents/IRONHACK/PROYECT/Final-Project/data/cleaned/players_matched'

all_files = os.listdir(folder_path)

csv_files = [f for f in all_files if f.endswith('.csv')]

overall_players_file_list = []

for csv in csv_files:
    file_path = os.path.join(folder_path, csv)
    try:
        # Try reading the file using default UTF-8 encoding
        df = pd.read_csv(file_path,index_col=0)
        overall_players_file_list.append(df)
    except UnicodeDecodeError:
        try:
            # If UTF-8 fails, try reading the file using UTF-16 encoding with tab separator
            df = pd.read_csv(file_path, sep='\t', encoding='utf-16',index_col=0)
            overall_players_file_list.append(df)
        except Exception as e:
            print(f"Could not read file {csv} because of error: {e}")
    except Exception as e:
        print(f"Could not read file {csv} because of error: {e}")
```

```python
Q1 = headshots_wlratio_revives_final['Kills'].quantile(0.25)
Q3 = headshots_wlratio_revives_final['Kills'].quantile(0.75)
IQR = Q3 - Q1

outliers = (headshots_wlratio_revives_final['Kills'] < Q1 - 1.5 * IQR) | (headshots_wlratio_revives_final['Kills'] > Q3 + 1.5 * IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to exclude outliers
headshots_wlratio_revives_final = headshots_wlratio_revives_final[(headshots_wlratio_revives_final['Kills'] >= lower_bound) & (headshots_wlratio_revives_final['Kills'] <= upper_bound)]
```
