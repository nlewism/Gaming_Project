```python
def player_statistics(driver):


    player_name = driver.find_element(By.CSS_SELECTOR,'#profile > div.trn-profile-header > div > div.trn-profile-header__main > h1 > span.trn-profile-header__name').text

    list_cards = driver.find_elements(By.CSS_SELECTOR,'.trn-card')

    dictionary_cards = {}
    for card in list_cards:
        card_name = card.find_element(By.CSS_SELECTOR,'.trn-card__header-title')
        data_stats = card.find_elements(By.CSS_SELECTOR,'div[data-stat]')

        dictionary_cards[card_name.text]={element.get_attribute('data-stat'):element.text for element in data_stats}
    dictionary_cards['name']=player_name
    return dictionary_cards
```

ATTACKER OPERATOR SPECIALTIES

```python
att_operator_attr = []

for link in att_operator_links:
    time.sleep(5)
    driver.get(link)
    operator_name = driver.find_element(By.CSS_SELECTOR, '.operator__header__icons__names').text
    specialty_attr = driver.find_elements(By.CSS_SELECTOR, '.operator__header__roles')
    
    if len(specialty_attr) < 3:
        att_list = [operator_name, specialty_attr[1].text]
    else:
        att_list = [operator_name, specialty_attr[1].text + specialty_attr[2].text]
    
    att_operator_attr.append(att_list)

specialty_df = pd.DataFrame(att_operator_attr, columns=['Operator Name', 'Attributes'])

has_comma = specialty_df['Attributes'].str.contains(',')

specialty_df['Attribute1'] = specialty_df.loc[has_comma, 'Attributes'].str.split(',').str[0]
specialty_df['Attribute2'] = specialty_df.loc[has_comma, 'Attributes'].str.split(',').str[1]

specialty_df.loc[~has_comma, 'Attribute1'] = specialty_df.loc[~has_comma, 'Attributes']
specialty_df['Attribute2'] = specialty_df['Attribute2'].fillna('')

specialty_df['Attribute1'] = specialty_df['Attribute1'].str.rstrip('\n')

unique_attributes = list(pd.unique(specialty_df[['Attribute1', 'Attribute2']].values.ravel()))
unique_attributes.remove('')


for attribute in unique_attributes:
    specialty_df[attribute] = 0


for attribute in unique_attributes:
    specialty_df.loc[(specialty_df['Attribute1'] == attribute) | (specialty_df['Attribute2'] == attribute), attribute] = 1

specialty_df.drop(['Attributes','Attribute1', 'Attribute2'], axis=1, inplace=True)
specialty_df_att = specialty_df
specialty_df_att
```

```python
att_operator_stats = {}


for link in att_operator_links:
    sub_class_list = []
    time.sleep(5)
    driver.get(link)
    operator_name = driver.find_element(By.CSS_SELECTOR, '.operator__header__icons__names').text
    child_class = driver.find_elements(By.XPATH, "//div[@class='react-rater']")
    
   
    for element in child_class:
        sub_elements = element.find_elements(By.XPATH, ".//*")
        
       
        for sub_element in sub_elements:
            sub_class = sub_element.get_attribute("class")
            sub_class_list.append(sub_class)
            
    
    att_operator_stats[operator_name] = sub_class_list


df = pd.DataFrame(att_operator_stats.items(), columns=['Operator Name', 'Stats'])

expanded_df = df['Stats'].apply(pd.Series)


expanded_df['Operator Name'] = df['Operator Name']


expanded_df = expanded_df[['Operator Name'] + [col for col in expanded_df.columns if col != 'Operator Name']]

columns_to_drop = [0,2,4,6,8,10,12,14,16]
filtered_df = expanded_df.drop(columns_to_drop, axis=1)

health_counts = (filtered_df.loc[:, [1, 3, 5]] == 'react-rater-star is-disabled is-active').sum(axis=1)
filtered_df['health'] = health_counts

speed_counts = (filtered_df.loc[:, [7, 9, 11]] == 'react-rater-star is-disabled is-active').sum(axis=1)
filtered_df['speed'] = speed_counts

difficulty_counts = (filtered_df.loc[:, [13, 15, 17]] == 'react-rater-star is-disabled is-active').sum(axis=1)
filtered_df['difficulty'] = difficulty_counts

columns_to_drop = [1, 3, 5, 7, 9, 11, 13, 15, 17]
filtered_att_df = filtered_df.drop(columns_to_drop, axis=1)
filtered_att_df
```

```python
legend_names = []
legend_specialty = []
legend_unique_ability = []

for link in legend_links:
   
    time.sleep(5)
    driver.get(link)

    try:
      
        name = driver.find_element(By.CSS_SELECTOR, 'body > main > ea-section:nth-child(2) > ea-section-column:nth-child(3) > ea-text > h1').text
    except NoSuchElementException:
        
        name = driver.find_element(By.CSS_SELECTOR, 'body > main > ea-section:nth-child(3) > ea-section-column:nth-child(3) > ea-text > h1').text
    
    legend_names.append(name)

    try:
       
        details_shadow = driver.find_element(By.CSS_SELECTOR, 'body > main > ea-section:nth-child(2) > ea-section-column:nth-child(3) > ea-details-table').shadow_root
    except NoSuchElementException:
      
        details_shadow = driver.find_element(By.CSS_SELECTOR, 'body > main > ea-section:nth-child(3) > ea-section-column:nth-child(3) > ea-details-table').shadow_root
    
   
    specialty = details_shadow.find_element(By.CSS_SELECTOR, 'div:nth-child(5)').text
    legend_specialty.append(specialty)

    
    unique_1 = details_shadow.find_element(By.CSS_SELECTOR, 'div:nth-child(6)').text
    unique_2 = details_shadow.find_element(By.CSS_SELECTOR, 'div:nth-child(7)').text
    unique_3 = details_shadow.find_element(By.CSS_SELECTOR, 'div:nth-child(8)').text
    legend_unique_ability.extend([unique_1, unique_2, unique_3])
```

```python
data = {
    'legend_name': legend_names,
    'legend_specialty': legend_specialty,
    'tactical_ability': legend_unique_ability[::3],  
    'passive_ability': legend_unique_ability[1::3],  
    'ultimate_ability': legend_unique_ability[2::3],  
}

legend_df = pd.DataFrame(data)


legend_df['tactical_ability'] = legend_df['tactical_ability'].str.replace('Tactical Ability', '')
legend_df['passive_ability'] = legend_df['passive_ability'].str.replace('Passive Ability', '')
legend_df['ultimate_ability'] = legend_df['ultimate_ability'].str.replace('Ultimate Ability', '')
legend_df['legend_specialty'] = legend_df['legend_specialty'].str.replace('Class ','')


legend_df['tactical_ability'] = legend_df['tactical_ability'].str.title()
legend_df['passive_ability'] = legend_df['passive_ability'].str.title()
legend_df['ultimate_ability'] = legend_df['ultimate_ability'].str.title()
```
