```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def calculate_counts(player_data):
    counts = {'ASSAULT': 0, 'RECON': 0, 'SUPPORT': 0, 'ENGINEER': 0}
    for col in player_data.index:
        if col != 'name' and player_data[col] == True:
            character = col
            specialty = specialty_mapping_df.loc[specialty_mapping_df['Character'] == character, 'Specialty'].iloc[0]
            counts[specialty] += 1
    return counts

# Apply the function to each row of the dataframe
headshots_wlratio_revives_no_outliers3['Counts'] = headshots_wlratio_revives_no_outliers3.apply(calculate_counts, axis=1)

# Convert the dictionary in the 'Counts' column to separate columns
headshots_wlratio_revives_no_outliers3 = pd.concat([headshots_wlratio_revives_no_outliers3.drop(['Counts'], axis=1), headshots_wlratio_revives_no_outliers3['Counts'].apply(pd.Series)], axis=1)

# Optionally, reorder columns if needed
headshots_wlratio_revives_no_outliers3 = headshots_wlratio_revives_no_outliers3[['name', 'ASSAULT', 'RECON', 'SUPPORT', 'ENGINEER']]
```
