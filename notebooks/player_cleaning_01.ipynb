# Overwatch

```python
folder_path = '/Users/samielzengotita/Documents/IRONHACK/PROYECT/Final-Project/data/raw/player_csvs/OVERWATCH PLAYERS'

all_files = os.listdir(folder_path)

csv_files = [f for f in all_files if f.endswith('.csv')]

overwatch_players_file_list = []

for csv in csv_files:
    file_path = os.path.join(folder_path, csv)
    try:
        # Try reading the file using default UTF-8 encoding
        df = pd.read_csv(file_path)
        overwatch_players_file_list.append(df)
    except UnicodeDecodeError:
        try:
            # If UTF-8 fails, try reading the file using UTF-16 encoding with tab separator
            df = pd.read_csv(file_path, sep='\t', encoding='utf-16')
            overwatch_players_file_list.append(df)
        except Exception as e:
            print(f"Could not read file {csv} because of error: {e}")
    except Exception as e:
        print(f"Could not read file {csv} because of error: {e}")
```

```python
overwatch_players_concatenated = pd.concat(overwatch_players_file_list, ignore_index=True)
overwatch_players_concatenated.drop(columns='Unnamed: 0',inplace=True)
overwatch_players_concatenated.drop_duplicates(subset=['name'], inplace=True)
```

```python
overwatch_players_concatenated['Top 3 Heroes'] = overwatch_players_concatenated['Top 3 Heroes'].astype(str).str.upper()
overwatch_heroes = pd.read_csv('/Users/samielzengotita/Documents/IRONHACK/PROYECT/Final-Project/data/raw/specialist_csvs/overwatch_heroes.csv')
overwatch_heroes = list(overwatch_heroes['Hero Name'])
for hero in overwatch_heroes:
    overwatch_players_concatenated[hero]= overwatch_players_concatenated['Top 3 Heroes'].apply(lambda x: hero in x)
overwatch_players_concatenated.drop(columns=['Top 3 Heroes','Cards','Environmental Kills','Multikills','Final Blows','Solo Kills','Objective Time','Hero Damage Done','Objective Kills','Offensive Assists','Defensive Assists','Recon Assists','Melee Final Blows'],inplace=True)
overwatch_players_concatenated = overwatch_players_concatenated.set_index('name').reset_index(level='name')

overwatch_players_concatenated['KDA Ratio'] = overwatch_players_concatenated['KDA Ratio'].apply(lambda x: x.replace(',', '') if isinstance(x, str) and ',' in x else x)
overwatch_players_concatenated['KDA Ratio'] = pd.to_numeric(overwatch_players_concatenated['KDA Ratio'], errors='coerce').astype(float)
overwatch_players_concatenated[overwatch_players_concatenated.select_dtypes(include='object').columns] = overwatch_players_concatenated.select_dtypes(include='object').apply(lambda x: x.str.replace(',', ''))
overwatch_players_concatenated[overwatch_players_concatenated.select_dtypes(include='object').columns] = overwatch_players_concatenated.select_dtypes(include='object').apply(pd.to_numeric, errors='ignore')
```

# Rainbow Six

```python
df = pd.read_csv('/Users/samielzengotita/Documents/IRONHACK/PROYECT/Final-Project/merged_rainbow.csv')
```

```python
unique_operators = pd.unique(df[['Top Operator 1', 'Top Operator 2', 'Top Operator 3']].values.ravel('K'))

for operator in unique_operators:
    df[operator] = 0

# Set values based on conditions
for index, row in df.iterrows():
    for operator in unique_operators:
        if row['Top Operator 1'] == operator or row['Top Operator 2'] == operator or row['Top Operator 3'] == operator:
            df.at[index, operator] = 1

df.drop(columns=['Top Operator 1', 'Top Operator 2', 'Top Operator 3','Unnamed: 0'],inplace=True)
```

```python
df.to_csv('r6_players_concatenated.csv')
```
