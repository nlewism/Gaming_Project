# R6 Top Player Stats

```python
def player_statistics(driver):


    player_name = driver.find_element(By.CSS_SELECTOR,'#profile > div.trn-profile-header > div > div.trn-profile-header__main > h1 > span.trn-profile-header__name').text

    list_cards = driver.find_elements(By.CSS_SELECTOR,'.trn-card')

    dictionary_cards = {}
    for card in list_cards:
        card_name = card.find_element(By.CSS_SELECTOR,'.trn-card__header-title')
        data_stats = card.find_elements(By.CSS_SELECTOR,'div[data-stat]')

        dictionary_cards[card_name.text]={element.get_attribute('data-stat'):element.text for element in data_stats}
    dictionary_cards['name']=player_name
    return dictionary_cards
```

# Apex Legends Top Players

```python
player_links = []

for i in range(1,101):
    player_link = driver.find_element(By.CSS_SELECTOR,f'#app > div.trn-wrapper > div.trn-container > div > main > div.leaderboards > div.content.mt-4 > div > div > div.board > div > table > tbody > tr:nth-child({i}) > td.username > div > a')
    x = player_link.get_attribute('href')
    player_links.append(x)
```

```python
list_cards = []
for link in player_links[42:]:
    #player_link = driver.find_element(By.CSS_SELECTOR,f'#app > div.trn-wrapper > div.trn-container > div > main > div.leaderboards > div.content.mt-4 > div > div > div.board > div > table > tbody > tr:nth-child({i}) > td.username > div > a')
    driver.get(link)
    time.sleep(5)
    #driver.refresh()
    
    dictionary_cards = {}
    try:
        player_name = driver.find_element(By.CLASS_NAME,'trn-ign__username').text

        mega_stats = driver.find_element(By.CLASS_NAME,'giant-stats').find_elements(By.CLASS_NAME,'wrapper')


#mega_tile_titles = []
#mega_tile_numbers = []

        for tile in mega_stats:

            mega_stat_title = tile.find_element(By.CLASS_NAME,'numbers').find_element(By.CLASS_NAME,'name').text
            mega_stat_number = tile.find_element(By.CLASS_NAME,'value').text
            dictionary_cards[mega_stat_title]=mega_stat_number

#        mega_tile_titles.append(mega_stat_title)
#        mega_tile_numbers.append(mega_stat_number)

        main_stats = driver.find_element(By.CLASS_NAME,'main').find_elements(By.CLASS_NAME,'numbers')

#main_stats_titles = []
#main_stats_numbers = []

        for tile in main_stats:
            main_stats_title = tile.find_element(By.CLASS_NAME,'name').text
            main_stats_number = tile.find_element(By.CLASS_NAME,'value').text
            dictionary_cards[main_stats_title]=main_stats_number
#        main_stats_titles.append(main_stats_title)
#        main_stats_numbers.append(main_stats_number)

        top_legends = driver.find_elements(By.CLASS_NAME,'legend__data')

        top_legend_list = []
        for legend in top_legends:
            top_legend = legend.find_element(By.CLASS_NAME,'legend__name').text
            top_legend_list.append(top_legend)
        dictionary_cards['Top 3 Legends']=top_legend_list
        dictionary_cards['name']=player_name
        list_cards.append(dictionary_cards)

    except NoSuchElementException:
        pass
    time.sleep(15)
```

```python
df = pd.json_normalize(list_cards)
```

```python
df.to_csv('apex_top_players_9.csv')
```
