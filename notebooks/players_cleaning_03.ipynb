```python
import pandas as pd
import os
```

```python
pd.set_option('display.max_columns', None)
```

```python
folder_path = '/Users/samielzengotita/Documents/IRONHACK/PROYECT/Final-Project/data/cleaned/player_csvs'

all_files = os.listdir(folder_path)

csv_files = [f for f in all_files if f.endswith('.csv')]

overall_players_file_list = []

for csv in csv_files:
    file_path = os.path.join(folder_path, csv)
    try:
        # Try reading the file using default UTF-8 encoding
        df = pd.read_csv(file_path,index_col=0)
        overall_players_file_list.append(df)
    except UnicodeDecodeError:
        try:
            # If UTF-8 fails, try reading the file using UTF-16 encoding with tab separator
            df = pd.read_csv(file_path, sep='\t', encoding='utf-16',index_col=0)
            overall_players_file_list.append(df)
        except Exception as e:
            print(f"Could not read file {csv} because of error: {e}")
    except Exception as e:
        print(f"Could not read file {csv} because of error: {e}")
```

```python
overwatch_players = overall_players_file_list[0]
```

```python
rainbow_players = overall_players_file_list[1]
```

```python
apex_players = overall_players_file_list[2]
```

```python
bf_players = overall_players_file_list[3]
```

```python
column_name_mapping = {
    'Eliminations': 'Kills',
    'Games Won': 'Matches_Won',
    'KDA Ratio': 'K/D'
}

overwatch_players.rename(columns=column_name_mapping, inplace=True)
```

```python
column_name_mapping = {
    'MatchesWon': 'Matches_Won',
    'KillAssists': 'Assists',
    'KDRatio': 'K/D'
}
columns_to_drop = ['MatchesLost','HoursPlayed']
rainbow_players.rename(columns=column_name_mapping, inplace=True)
rainbow_players.drop(columns=columns_to_drop,inplace=True)
```

```python
rainbow_players.head()
```

```python
column_name_mapping = {
    'Wins': 'Matches_Won',
}
apex_players['WLRatio'] = (apex_players['Wins'] / (apex_players['Matches Played'] - apex_players['Wins']))*100

apex_players.rename(columns=column_name_mapping, inplace=True)
columns_to_drop = ['Matches Played']
apex_players.drop(columns=columns_to_drop,inplace=True)
```

```python
apex_players.head()
```

```python
column_name_mapping = {
    'Headshot_Kills': 'Headshots',
    'Win_Ratio': 'WLRatio',
}
columns_to_drop = ['Matches_Lost','Multi_Kills','Melee_Kills']

bf_players.rename(columns=column_name_mapping, inplace=True)
bf_players.drop(columns=columns_to_drop,inplace=True)
bf_players.rename(columns={'Gamer_Tag': 'name'}, inplace=True)
```

```python
column_order_df1 = ['name', 'Kills', 'Matches_Won', 'Deaths', 'Assists', 'K/D','Game']
column_order_df2 = ['name', 'Kills', 'Matches_Won', 'Deaths', 'Assists', 'K/D', 'Headshots', 'WLRatio', 'Revives','Game']
column_order_df3 = ['name', 'Kills', 'Matches_Won', 'Headshots', 'WLRatio', 'Revives','Game']
column_order_df4 = ['name', 'Kills', 'Matches_Won', 'Deaths', 'Assists', 'K/D', 'Headshots', 'WLRatio', 'Revives','Game']

# Reorder columns and include remaining columns for each DataFrame
overwatch_players = overwatch_players[column_order_df1 + [col for col in overwatch_players.columns if col not in column_order_df1]]
rainbow_players = rainbow_players[column_order_df2 + [col for col in rainbow_players.columns if col not in column_order_df2]]
apex_players = apex_players[column_order_df3 + [col for col in apex_players.columns if col not in column_order_df3]]
bf_players = bf_players[column_order_df4 + [col for col in bf_players.columns if col not in column_order_df4]]
```
